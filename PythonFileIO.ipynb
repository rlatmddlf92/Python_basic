# 파일 처리
|파일열기모드|	설명|
|-------|--------|
|r	|읽기모드 - 파일을 읽기만 할 때 사용|
|w	|기모드 - 파일에 내용을 쓸 때 사용|
|a	|추가모드 - 파일의 마지막에 새로운 내용을 추가 시킬 때 사용|
|b|	2진 모드(바이너리 모드)|
|t|	텍스트 모드 (기본값)|

```python
f = open("a.txt", 'w')
# f=open('a.txt', mode='wt', encoding='utf-8')
f.write('this first line\n')   # write a line to the file
f.write('this second line\n')  # write one more line to the file
f.close()
```

```python
f = open('a.txt', 'a')
f.write("this is third line\n")
f.close()
```

```python
f = open("b.txt", 'w')
for i in range(10):
    f.write('this '+str(i)+' line\n')  # write one more line to the file
f.close()
```

```python
import os
os.remove("a.txt")
```

```python
f = open('b.txt', 'r')
for line in f:
    print(line)
```

# pickle을 통한 파일 입출력

```python
import pickle
f = open('pickle.dat', 'wb') #wb: write binary
pickle.dump(11, f)
pickle.dump("this is a line", f)
pickle.dump([1, 2, 3, 4], f)
f.close()
```

```python
import pickle
f = open('pickle.dat', 'rb') #rb: read binary
print(pickle.load(f))
print(pickle.load(f))
print(pickle.load(f))
f.close()
```

# shelve를 통한 자료 객체 입출력 
* shelve 라이브러리에서 제공하는 shelve는 key - value 형식으로 동작하는 JSON, dictionary 와 같이 동작하는 자료 객체

```python
import shelve
shelfFile = shelve.open('mydata')
cats = ['Zophie', 'Pooka', 'Simon']
shelfFile['cats'] = cats
shelfFile['animal'] = ['tiger','lion']
shelfFile['testDict'] = {'key':'value','key1':'value2'}
shelfFile.close()
```
